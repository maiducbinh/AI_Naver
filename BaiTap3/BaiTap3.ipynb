```python
# library
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import json
from sklearn.preprocessing import LabelEncoder
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("train_PTIT.csv")
df
```

```python
# preprocessing subject

#data_sub = df["subject"].unique()
name_2_Id = {
    "Probabilistic_Methods": 0,
    "Neural_Networks": 1,
    "Genetic_Algorithms": 2,
    "Rule_Learning": 3,
    "Reinforcement_Learning": 4,
    "Case_Based": 5,
    "Theory": 6,
}

df["subject"] = df["subject"].map(name_2_Id)
df

# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, :-1], df.iloc[:, -1], test_size=0.2, random_state=42
)
```

```python
# Decision tree

# library
from sklearn.tree import DecisionTreeClassifier

# fit
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
# predict
y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for Decision Tree:", accuracy)
```

```python
# Naive Bayes

# library
from sklearn.naive_bayes import GaussianNB

# fit
nb = GaussianNB()
nb.fit(X_train, y_train)

# predict
y_pred = nb.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for Naive Bayes:", accuracy)
```

```python
# SVM 

# library
from sklearn.svm import SVC

# fit 
svm_linear = SVC(kernel='linear') # sử dụng phép toán tích vô hướng giữa hai vector để tính toán độ tương đồng giữa chúng.
svm_linear.fit(X_train, y_train)

svm_poly = SVC(kernel='poly') # sử dụng hàm đa thức
svm_poly.fit(X_train, y_train)

svm_rbf = SVC(kernel='rbf') # sử dụng hàm hạt nhân Gaussian
svm_rbf.fit(X_train, y_train)

svm_sigmoid = SVC(kernel='sigmoid') # sử dụng hàm sigmoid
svm_sigmoid.fit(X_train, y_train)

# predict
y_linear_pred = svm_linear.predict(X_test)
y_poly_pred = svm_poly.predict(X_test)
y_rbf_pred = svm_rbf.predict(X_test)
y_sigmoid_pred = svm_sigmoid.predict(X_test)


print("Accuracy for SVM if kernel = linear:", accuracy_score(y_test, y_linear_pred))

print("Accuracy for SVM if kernel = poly:", accuracy_score(y_test, y_poly_pred))

print("Accuracy for SVM if kernel = rbf:", accuracy_score(y_test, y_rbf_pred))

print("Accuracy for SVM if kernel = sigmoid:", accuracy_score(y_test, y_sigmoid_pred))
```
